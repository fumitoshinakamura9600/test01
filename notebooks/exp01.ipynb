```python
# Sine plot
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x)

plt.figure(figsize=(8,4))
plt.plot(x, y, label='sin(x)')
plt.title('Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()
plt.savefig('sine_plot.png', dpi=150)
print('Saved: sine_plot.png')
plt.show()
```
